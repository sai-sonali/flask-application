```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
audio_file = "GPS _school_voice_samples srinivas_g22.wav"
```

```python
ipd.Audio(audio_file)
```

```python
# load audio files with librosa
signal, sr = librosa.load(audio_file)
```

```python
signal.shape
```

```python
#extracting MFCC's
mfccs = librosa.feature.mfcc(y=signal, n_mfcc=39, sr=sr)
```

```python
mfccs.shape
```

```python
fr=len(mfccs[0])
fr
```

```python
df=pd.read_csv('srinivas_g22.csv')
df.columns
```

```python
GXmfcc=[]
mfcc=mfccs.transpose()
mfcc=mfcc.tolist()
#print(len(mfcc[38]))
GXmfcc.extend(mfcc)
```

```python
stime=list(df['Start Time'])
etime=list(df['End Time '])
code=list(df['Class'])
print(len(code))
for i in range(len(code)):
    if(code[i]>0):
        code[i]=1
```

```python
# ad=11.610045
ad=8.77960
t=ad/fr
t1=0
t2=ad/fr
y=[]
l=len(stime)
i=0
cnt=0
while(t2<ad and t1<=t2):
    if(i<l-1):
        while(t1<stime[i] and t2<stime[i]):
            y.append(0)
            t1+=t
            t2+=t
        while((t1>=stime[i] and t2<=etime[i]) or (t1<stime[i] and t2>=stime[i]) or (t1<etime[i] and t2>etime[i] and t2<stime[i+1])):
            y.append(code[i])
            t1+=t
            t2+=t
        i+=1
    else:
        y.append(0)
        t1+=t
        t2+=t

print(cnt)
print(len(y))
print(y)
print(y.count(1))
```

```python
X=pd.DataFrame(GXmfcc)
Y=pd.DataFrame(y)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3,random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=None)
clf.fit(X_train,Y_train)
```

```python
# Use the trained classifier to predict the labels of the test set
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate accuracy
accuracy = accuracy_score(Y_test, y_pred)

# Calculate precision
precision = precision_score(Y_test, y_pred)

# Calculate recall
recall = recall_score(Y_test, y_pred)

# Calculate F1 score
f1 = f1_score(Y_test, y_pred)

# Print the results
print("Accuracy: ", accuracy*100)
print("Precision: ", precision)
print("Recall: ", recall)
print("F1 Score: ", f1)
print(y_pred)
# Use the trained classifier to predict the class probabilities of the test set
y_proba = clf.predict_proba(X_test)

# Print the class probabilities
```
